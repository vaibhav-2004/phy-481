```python
## general setup for the plot

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 6, 0.5)
y = np.arange(-5, 6, 0.5)
X, Y = np.meshgrid(x, y)
```

```python
## 1) 
U = X
V = Y

plt.figure(figsize=(6,6))
plt.quiver(X,Y,U,V)
plt.xlabel("x")
plt.ylabel("y")
plt.title("vector field")
plt.axis("equal")
plt.show()

## 5) 
## Arrows point radially outward and grow with distance. 
## Represents expansion from a source, like an exploding gas cloud
```

```python
## 2) v(x,y) = r^/r

R = np.sqrt(X**2 + Y**2)
U = X / (R**2)
V = Y / (R**2)

plt.figure(figsize=(6,6))
plt.quiver(X,Y,U,V)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.title("vector field")
plt.axis("equal")
plt.show()

## 5) 
## Radial field wakens as 1/r.
## It can describe the electric field of a line charge of a point source.
```

```python
## 3) v = r(vec)/r^3

R = np.sqrt(X**2 + Y**2)
U = X / (R ** 3)
V = Y / (R ** 3)

plt.figure(figsize=(6,6))
plt.quiver(X,Y,U,V)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.title("vector field")
plt.axis("equal")
plt.show()

## 5)
## Radial field weakens as 1/r^2
## It can describe the coloumb or gravitaional field of a point source.
```

```python
## 4) v(x,y) = $/phi$

R = np.sqrt(X**2+Y**2)
U = -Y / R
V = X / R

plt.figure(figsize=(6,6))
plt.quiver(X,Y,U,V)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.title("vector field")
plt.axis("equal")
plt.show()

## 5)
## Tangential field rotating around the origin with equal strength.
## It can be described as the magnetic field around a staright current carrying wire.
```
