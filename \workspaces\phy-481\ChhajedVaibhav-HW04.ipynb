```python
## 2) 
import numpy as np
import matplotlib.pyplot as plt

# Constants
k = 9e9
q = 1.0
r0 = np.array([0, 0])  # charge at origin


x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, y)

Ex = np.zeros_like(X)
Ey = np.zeros_like(Y)

for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        r = np.array([X[i, j], Y[i, j]]) - r0
        r_mag = np.linalg.norm(r)
        if r_mag != 0:
            E = (k*q*r) / r_mag**3
            Ex[i, j], Ey[i, j] = E[0], E[1]

# Plot
plt.figure(figsize=(6, 6))
plt.quiver(X, Y, Ex, Ey)
plt.scatter(r0[0], r0[1], color="red", s=80, label="Charge q")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Electric Field of a Point Charge")
plt.axis("equal")
plt.legend()
plt.show()
```

```python
## 3) 
import numpy as np
import matplotlib.pyplot as plt

# Constants
k = 9e9
q = 1.0
d = 1.0  
pos_plus = np.array([0, d/2])
pos_minus = np.array([0, -d/2])

x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)

Ex = np.zeros_like(X)
Ey = np.zeros_like(Y)


for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        r = np.array([X[i, j], Y[i, j]])

       
        r_plus = r - pos_plus
        r_plus_mag = np.linalg.norm(r_plus)
        if r_plus_mag != 0:
            E_plus = (k * q * r_plus) / r_plus_mag**3
        else:
            E_plus = 0

        
        r_minus = r - pos_minus
        r_minus_mag = np.linalg.norm(r_minus)
        if r_minus_mag != 0:
            E_minus = (-q * k  * r_minus) / r_minus_mag**3
        else:
            E_minus = 0

        E = E_plus + E_minus
        Ex[i, j], Ey[i, j] = E[0], E[1]

# Plot
plt.figure(figsize=(6, 6))
plt.quiver(X, Y, Ex, Ey)
plt.scatter(pos_plus[0], pos_plus[1], color="red", s=80, label="+q")
plt.scatter(pos_minus[0], pos_minus[1], color="blue", s=80, label="−q")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Electric Field of a Dipole")
plt.axis("equal")
plt.legend()
plt.show()


```
